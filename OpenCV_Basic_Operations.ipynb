# Grayscaling Image
In a normal image, it is made up of three layers, Red,Blue,Green and thus forms a three dimensional matrix while storing, but in grayscale form it is two dimensional matrix and is thus compressed.

```python
grayimg=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(grayimg)
```

```python
print(grayimg.shape)
```

# Translation
Translation is the process of shifiting image from one location to another location.
This can be done using transformation matrix.

```python
# Storing height and width of the image 
h, w = img.shape[:2]
```

```python
qh, qw = h / 4, w / 4
```

```python
import numpy as np
T = np.float32([[1, 0, qw], [0, 1, qh]]) 
# We use warpAffine function to transform 
# the image using the matrix, T 
img_translation = cv2.warpAffine(img, T, (w, h)) 
  

cv2_imshow(img_translation) 
```

# Edge Detection
It involves the detecting the sharp edges in the image.

```python
edges = cv2.Canny(img, 100, 200) 
# syntax for Canny function Python

# cv2.Canny(image, threshold1, threshold2, apertureSize, L2gradient
cv2_imshow(edges)
```

```python
edges=cv2.Canny(img,100,150)
cv2_imshow(edges)
# threshold1: It is the High threshold value of intensity gradient.
# threshold2: It is the Low threshold value of intensity gradient
```

# Sharpening
This operation is used to sharpen the image

cv2.filter2D() this fucntion is used to perform the sharpen/blur opeeration. It has 3 parameters in which one is kernal. kernal is nothing but the numpy array for sharpning image.

```python
#creating kernel
kernel_sharpening = np.array([[-1,-1,-1], 
                              [-1, 9,-1],
                              [-1,-1,-1]])
# applying the sharpening kernel to the input image & displaying it.
sharpened = cv2.filter2D(img, -1, kernel_sharpening)
cv2_imshow(sharpened)
```

```python
#creating kernel
kernel_sharpening = np.array([[0,-1,0], 
                              [-1, 4,-1],
                              [0,-1,0]])
# applying the sharpening kernel to the input image & displaying it.
sharpened = cv2.filter2D(img, -1, kernel_sharpening)
cv2_imshow(sharpened)
```
